## Question 1:  Is the data normal?

Answer: Yes.  A visual inspection varying bin size shows only a slight difference in the shape of the left and right tails.  Additionally, the qqplot (quintile-quintile) plot and the scipy test for normality both indicate a normal distribution.


```python
# q-q plot
qqplot(df.temperature, line='s')
plt.show()

# normality test
stat, p = normaltest(df.temperature)
print 'Statistics= {:.3f}, p= {:.3f}'.format(stat, p)
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

## Question 2: Is the sample size large? Are the observations independent?

There are 65 male and 65 female observations for a total of 130.  For the population as a whole, with a 95% confidence, the mean human body temperature will be between 98.1 and 98.4, and we can conclude that the sample is relatively large.

The description of the measurement methodology is not given, so I assume each observation is a separate subject (rather than multiple measurements of the same person) and that the instrumentation is well calibrated and in good working order.  With these conditions, the data can be assumed to be independent and the CLT holds true.


```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

mean_confidence_interval(df.temperature)
```

## Q3:  Is the true population mean really 98.6 degrees F?

The null hypothesis in this case is that the population mean temperature is not 98.6 degreesF.  As the cell below shows, the overall sample population has a mean temp of 98.25, yielding a z score of 5.48 which is a pvalue of <0.02 and we reject the null hypothesis. 

It is very unlikely that the true population mean is 98.6.  For women the z score is lower, but still yields a pvalue of approximately 0.01, rejecting the null hypothesis for females as well as the overall population.


```python
def zscore(target_mean,data):
    '''data is received in np array'''
    sample_mean=np.mean(data)
    zscore=(target_mean-sample_mean)/(np.std(data)/np.sqrt(len(data)))
    return zscore

print "Mean temp for total pop is %5.2f"%df.temperature.mean()," with a Zscore of",zscore(98.6,df.temperature)
print "Mean temp for women is %5.2f"%df.loc[(df.gender=='F'),"temperature"].mean()," with a Zscore of",zscore(98.6,df.loc[(df.gender=='F'),"temperature"])
print "Mean temp for men is %5.2f"%df.loc[(df.gender=='M'),"temperature"].mean()," with a Zscore of",zscore(98.6,df.loc[(df.gender=='M'),"temperature"])
    
```

## Q4)  Draw a small sample of size 10 from the data and repeat both tests.
The ztest for women gives lower but still confidence hat the same mean is not 98.6 degrees.  For women it is 91.6% and men it is 94.2% confidence.

The T test shows a 90% confidence for women and something between 90% and 95% confidence for men that the mean is not 98.6 for the sample.  In this case, the T test is more appropriate because the number of observations is so small.


```python
#Split DF into gender
just_women_df=df.loc[(df.gender=='F'),"temperature"]
just_men_df=df.loc[(df.gender=='M'),"temperature"]

women_resampled=np.empty(10)
men_resampled=np.empty(10)

women_resampled=np.random.choice(just_women_df,10)
men_resampled=np.random.choice(just_men_df,10)

print "Resampled case, 10 observations each gender"
print "Mean temp for women is %5.2f"%np.mean(women_resampled)," with a Zscore of %5.2f"%zscore(98.6,women_resampled)
print "Mean temp for men is %5.2f"%np.mean(men_resampled)," with a Zscore of %5.2f"%zscore(98.6,men_resampled)
```

## Q5) At what temperature should we consider someone's temperature to be "abnormal"?

For discussion, assume an abnomal temperature would be outside the standard whiskers on a Box and Whiskers plot, or, +/- 1.5 * IQR.  The cell below calculates appropriate highs and lows for women and men.  For simplicity, women's temperatures should be considered abnormal if outside the range: 97.2-99.6 and men outside the range 96.6-99.6.  It is also important to note that the shape of the tails was different, and the the female population had a larger number of outliers
 

```python
women_mean=np.mean(just_women_df)
women_std=np.std(just_women_df)
print "Women mean: %5.1f" % women_mean, "wih a standard deviation of: %5.1f" % women_std
print "Women IQR:",np.percentile(just_women_df,[25,75])
iqr_temp=np.percentile(just_women_df,75)-np.percentile(just_women_df,25)
print "Women abnormal high: %5.1f" % (women_mean+1.5*iqr_temp)
print "Women abnormal low: %5.1f" % (women_mean-1.5*iqr_temp)

men_mean=np.mean(just_men_df)
men_std=np.std(just_women_df)
print "\nMen mean: %5.1f" % men_mean, "wih a standard deviation of: %5.1f" % men_std
print "Men IQR:",np.percentile(just_men_df,[25,75])
miqr_temp=np.percentile(just_men_df,75)-np.percentile(just_men_df,25)
print "Men abnormal high: %5.1f" % (men_mean+1.5*miqr_temp)
print "Men abnormal low: %5.1f" % (men_mean-1.5*miqr_temp)

sns.boxplot(x="gender",y="temperature",data=df)
plt.title("Test subject temperatures")
plt.show()
```

## Q6) Is there a significant difference between males and females in normal temperature?

In testing 130 subjects, half men and half women, women's mean temperture was slightly higher than men (by 0.3F), and had ouliers on the high side of the distribution.  The null hypothesis is that the mean body temperatures are the same.  The ttest returns a pvalue of 0.02 and the null hypothesis is rejected.  It is very unlikely that men and women have the same mean body temperature.

```python
print stats.ttest_ind(just_men_df,just_women_df)
sns.distplot(just_men_df,label='men')
sns.distplot(just_women_df,label='women')
plt.legend(loc='best')
plt.title('Distributions of men and women')
plt.show()
```

```python
# Generate single image for summary discussion above
def gen_combination_plot():
    '''Combine two main plots from excercise'''
    c1= (0.29803921568627451, 0.44705882352941179, 0.69019607843137254)
    c2= (0.33333333333333331, 0.6588235294117647, 0.40784313725490196)

    plt.figure(figsize=(10,4),dpi=90)
    plt.subplot(1,2,1)

    palette=[c1,c2]
    sns.boxplot(x="gender",y="temperature",data=df,palette=palette)


    plt.subplot(1,2,2)

    bins = np.linspace(df.temperature.min()-1, df.temperature.max()+1, 9)

    plt.hist(just_women_df,bins,alpha=0.85, label='Women',color=c1)
    plt.hist(just_men_df,bins,alpha=0.65, label='Men',color=c2)
    plt.xlabel("Measured Temperature")
    plt.ylabel("Count")
    plt.legend(loc='best')

    plt.suptitle("130 Total Test Subjects",fontsize=14)
    plt.show()
    return
```
